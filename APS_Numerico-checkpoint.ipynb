#### Dependências

```r
library(polynom)
library(Orcs)
library(sp)
library(geosphere)
require(splancs, warn.conflicts = FALSE)
```

#### Lendo os dados de Logintude e Latitude dos pontos

```r
x <- read.csv("lago.csv",header=T)$Longitude
y <- read.csv("lago.csv",header=T)$Latitude
```

```r
for(i in 1:(length(x)-1))
{
    if(x[i]==x[i+1])
    {  
        x[i+1]=x[i+1]+0.000000001
    } 
}
```

```r
for(i in 1:(length(y)-1))
{
    if(y[i]==y[i+1])
    {  
        y[i+1]=y[i+1]+0.000000001
    } 
}
```

```r
plot(x,y,xlim=c(range(x)),ylim=c(range(y)))
```

```r
plot(x,y,xlim=c(range(x)),ylim=c(range(y)),type="l")

    for(i in 1:(length(x)-1))
    {
        a<-poly.calc(c(x[i],x[i+1]),c(y[i],y[i+1]))
        a1=a[1]
        a2=a[2]
        f<-function(x){a1+a2*x}
        plot(f,x[i],x[i+1],xlim=c(range(x)),ylim=c(range(y)),add=T, type="l")
    }
```

### Convertendo as coordenadas em um polígono e plotando-o em seguida

```r
coords_long_lat <- cbind(c(x), c(y))

poly_lago <- coords2Polygons(coords_long_lat, ID="Lago_1")

plot(poly_lago, col = "blue",xlim=c(range(x)),ylim=c(range(y)))
```

#### Área do polígono acima (em m²)

```r
areaPolygon(poly_lago)
```

### Lendo os dados convertidos em UTM

Foi utilizado o site **mygeodata.cloud** para a conversão das coordenadas

```r
x_utm <- read.csv("lago_utm.csv", header=T)$X
y_utm <- read.csv("lago_utm.csv", header=T)$Y
```

```r
#medida para tratar possíveis pontos duplicados

for(i in 1:(length(x_utm)-1))
{
    if(x_utm[i]==x_utm[i+1])
    {  
        x_utm[i+1]=x_utm[i+1]+0.000000001
    } 
}

for(i in 1:(length(y_utm)-1))
{
    if(y_utm[i]==y_utm[i+1])
    {  
        y_utm[i+1]=y_utm[i+1]+0.000000001
    } 
}
```

```r
plot(x_utm,y_utm,xlim=c(range(x_utm)),ylim=c(range(y_utm)))
```

```r
plot(x_utm,y_utm,xlim=c(range(x_utm)),ylim=c(range(y_utm)),type="l")

    for(i in 1:(length(x_utm)-1))
    {
        a<-poly.calc(c(x_utm[i],x_utm[i+1]),c(y_utm[i],y_utm[i+1]))
        a1=a[1]
        a2=a[2]
        f<-function(x_utm){a1+a2*x_utm}
        plot(f,x_utm[i],x_utm[i+1],xlim=c(range(x_utm)),ylim=c(range(y_utm)),add=T, type="l")
    }
```

```r
coords_utm <- cbind(c(x_utm), c(y_utm))
areapl(coords_utm)
```

### Comparando as áreas obtidas nos dois métodos

```r
#area (em m²) obtida com a conversão dos pontos para UTM
areapl(coords_utm)
```

```r
#area (em m²) obtida do polígono criado através das coordenadas geográficas
areaPolygon(poly_lago)
```

```r
abs(areapl(coords_utm) - areaPolygon(poly_lago))
```
